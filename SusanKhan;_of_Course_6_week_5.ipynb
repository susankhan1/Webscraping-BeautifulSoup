```python
#List all case numbers for crimes involving minors?
%sql SELECT * FROM ChicagoCrimeData Where lower(Description) like '%minor%' or Description like '%child%' or \
Description like '%kid' or Description like '%infant'
```

```python
#List all kidnapping crimes involving a child?
%sql select * from ChicagoCrimeData where description like '%kidnap%' or description like '%abduction%';
```

```python
#List Problem 5: What kind of crimes were recorded at schools?
%sql select primary_type, location_description from ChicagoCrimeData where location_description like '%school%';
```

```python
#Problem 6: List the average safety score for all types of schools.
#possibility: %sql select AVG(safety_score) as Average_Safety_Score from chicagopublicschools;
%sql select elementarymiddleorhighschool, AVG(safety_score) from chicagopublicschools group by elementarymiddleorhighschool;
```

```python
#Problem 7: List 5 community areas with highest % of households below poverty line.
%sql select Community_Area_Name, percent_households_below_poverty from ChicagoCensusData order by percent_households_below_poverty desc LIMIT 5;
```

```python
#Problem 8: Which community area(number) is most crime prone?
%sql select community_area_number as Community_No, Community_Area_Name as Area, COUNT(community_area_number) as frequency from ChicagoCensusData order by community_area_number DESC LIMIT 1;
```

```python
#problem 9: Use a sub-query to find the name of the community area with highest hardship index
#possibility: %sql select community_area_number as NO, Community_Area_Name as Name, hardship_index as hardship from ChicagoCensusData order by hardship_index DESC LIMIT 1;
%sql select community_area_number as NO, Community_Area_Name as Name, hardship_index as hardship from ChicagoCensusData where hardship_index = (select max(hardship_index) from ChicagoCensusData);

```

```python
#Problem 10: Use a sub-query to determine the Community Area Name with most number of crimes?
%sql select Community_Area_Name as Area from ChicagoCensusData where community_area_number = (select community_area_number from ChicagoCrimeData group by community_area_number order by count(community_area_number) DESC LIMIT 1) LIMIT 1;
```
